```python
import os
import pandas as pd
PATH='models'
def obtener_ultimo_periodo(pais, archivo_csv):
    ruta_pais = os.path.join('D:\Projects\data analytics\models', pais)
    ruta_pais = ruta_pais.replace("\\", "/")
    periodos = [x for x in os.listdir(ruta_pais) if os.path.isdir(os.path.join(ruta_pais, x))]
    ultimo_periodo = sorted(periodos)[-1]
    ruta_archivo = os.path.join(ruta_pais, ultimo_periodo, archivo_csv)
    data=pd.read_csv(ruta_archivo,delimiter=';')
    return data
```
